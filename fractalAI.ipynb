```python
import numpy as np
import matplotlib.pyplot as plt

def julia_set(c, width, height, x_min, x_max, y_min, y_max, max_iterations):
    # Create a grid of complex numbers
    x = np.linspace(x_min, x_max, width)
    y = np.linspace(y_min, y_max, height)
    Z = np.array(np.meshgrid(x, y)).T.reshape(-1, 2)
    Z = Z[:, 0] + 1j * Z[:, 1]
    
    # Initialize the output array
    img = np.zeros(Z.shape, dtype=int)
    
    # Iterate over each point in the grid
    for i in range(max_iterations):
        mask = np.abs(Z) < 2  # Only iterate on points inside the set
        img[mask] = i
        Z[mask] = Z[mask] ** 2 + c  # Update the points

    return img.reshape(height, width)

def plot_julia_set(image, c):
    plt.imshow(image, cmap='inferno', extent=(-2, 2, -2, 2))
    plt.title(f"Julia Set for c = {c}")
    plt.axis('off')
    plt.colorbar()
    plt.show()

# User input for Julia set parameter
real_part = float(input("Enter the real part of c (e.g., -0.4): "))
imaginary_part = float(input("Enter the imaginary part of c (e.g., 0.6): "))
c = complex(real_part, imaginary_part)

# Parameters for the plot
width, height = 800, 800
max_iterations = 100
x_min, x_max = -2, 2
y_min, y_max = -2, 2

# Generate and plot the Julia set
julia_image = julia_set(c, width, height, x_min, x_max, y_min, y_max, max_iterations)
plot_julia_set(julia_image, c)
```
